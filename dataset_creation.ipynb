```python
import csv
```

```python
sentences = []
sentiments = []
count = 0
with open ("data.csv ", "r", encoding="utf8") as f:
    reader = csv . reader ( f )
    # ignore the first line
    next ( reader )
    for row in reader :
        count += 1
        sentences . append ( row[0])
        sentiments . append ( row[1])
        # if count == 4:
        #     break

# print(sentences)
# print(sentiments)
```

```python
word_tokenized_sentences = [ sentence.split() for sentence in sentences ]
print(word_tokenized_sentences [:2])
```

```python
max_len = max ([len ( sentence ) for sentence in word_tokenized_sentences ])
# max_len = 64
padded_sentences = [ sentence + ['[PAD ]'] * max(0,max_len - len(sentence)) for sentence in word_tokenized_sentences ]
print ( padded_sentences [:2])
```

```python
vocab = {}
vocab ['[PAD]'] = 0
for sentence in padded_sentences :
    for token in sentence :
        if token not in vocab :
            vocab [ token ] = len( vocab )

print ( vocab )
```

```python
# convert the tokens to ids
padded_sentences_ids = [[ vocab [ token ] for token in sentence ] for sentence in padded_sentences ]
print ( padded_sentences_ids [:2])
```

```python
from transformers import DistilBertTokenizer
model_ckpt = "distilbert-base-uncased"
distilbert_tokenizer = DistilBertTokenizer . from_pretrained(model_ckpt)

bert_tokenized_sentences_ids = [ distilbert_tokenizer.encode(sentence, padding ='max_length', truncation =True , max_length = max_len)for sentence in sentences ]
print ( bert_tokenized_sentences_ids [:2])

bert_tokenized_sentences_tokens = [ distilbert_tokenizer.convert_ids_to_tokens(sentence ) for sentence in bert_tokenized_sentences_ids]
print ( bert_tokenized_sentences_tokens [:2])
```

```python
from transformers import DistilBertModel
import torch

model_name = 'distilbert/distilbert-base-uncased'
distilbert_model = DistilBertModel . from_pretrained ( model_name )

word_embeddings = []
```

```python
# convert padded sentence tokens into ids
for tokens in padded_sentences_ids :
    input_ids = torch . tensor ( tokens ) . unsqueeze ( 0 )
    with torch . no_grad ():
        outputs = distilbert_model ( input_ids )
    word_embeddings . append ( outputs . last_hidden_state )

print ( word_embeddings [0]. shape )

subword_embeddings = []
for tokens in bert_tokenized_sentences_ids :

    input_ids = torch . tensor ( tokens ) . unsqueeze ( 0 )
    with torch . no_grad ():
        outputs = distilbert_model ( input_ids )
    subword_embeddings . append ( outputs . last_hidden_state )

print ( subword_embeddings [0]. shape )
```

```python
embedding_classes = []
for i,embedding in enumerate(word_embeddings):
    if sentiments[i] == 'positive':
        embedding_vector = torch.Tensor([1,0,0])
    elif sentiments[i] == 'neutral':
        embedding_vector = torch.Tensor([0,1,0])
    elif sentiments[i] == 'negative':
        embedding_vector = torch.Tensor([0,0,1])
    # embedding_dict['sentence'] = embedding
    # embedding_dict['sentiment'] = embedding_vector
    # embedding_classes.append(embedding_dict)
    embedding_classes.append({'sentence': embedding,'sentiment': embedding_vector})
    # print(embedding)
    # if i == 2:
    #     break

# print(embedding_classes)
```

```python
print(embedding_classes[0]['sentiment'].shape)

list_length = len(embedding_classes)
split = int(0.8*list_length)

divided = {}

divided['train'] = embedding_classes[:split]
divided['test'] = embedding_classes[split:]
```

```python
with open('custom_embedding_classes.pkl', 'wb') as handle:
    pickle.dump(divided, handle, protocol=pickle.HIGHEST_PROTOCOL)
```
